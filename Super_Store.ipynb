# preprocessing

```python
#error  = 
df["Order Date"] = pd.to_datetime(df["Order Date"]  , format = 'mixed')
df["Ship Date"] = pd.to_datetime(df["Ship Date"] , format = 'mixed' )
```

```python
df.info()
```

```python
df["month_name"] = df["Order Date"].dt.month_name()
```

# monthly Sales 

```python
monthly_sales = df.groupby("month_name")['Sales'].sum().reset_index()
```

```python
#insights : check for Sales trend and peaks
```

# Monthly_sale highest in novermber(271461)

```python
print(monthly_sales)
```

# visualization
Line chart are use

```python
plt.figure(figsize=(8,5))
sns.lineplot(data = monthly_sales , x = 'month_name' , y = 'Sales', marker='o')
plt.title('Monthly sales')
plt.xlabel('month_name')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)

plt.savefig("month_sale.png" , dpi = 300)
plt.show()
```

# visualization monthly by profit

```python
monthly_profit = df.groupby('month_name')['Profit'].sum().reset_index()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data = monthly_profit  , x = 'month_name' , y = 'Profit' , marker = 'o')
plt.xlabel('month_name')
plt.ylabel('Profit')
plt.title('Monthly Profit')
plt.grid(True)
plt.xticks(rotation=45)
plt.savefig("month_profit.png" , dpi = 300)
plt.show()
```

# Highest Profit IN Month September(36045)

```python
import matplotlib.pyplot as plt
import seaborn as sns

category_sales = df.groupby('Category')['Sales'].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=category_sales, x='Category', y='Sales', palette=['r', 'g', 'b'],legend = False)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()


```

```python
category_profit = df.groupby('Category')['Profit'].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=category_profit, x='Category', y='Profit', palette=['r', 'g', 'b'])
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Profit by Category')
plt.show()
```

```python
profit_segment = df.groupby('Segment')['Profit'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data = profit_segment,x ='Segment',y='Profit',palette=('Set1'))
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.title('Profit by Segment')
plt.show()
```

```python
sales_segment = df.groupby('Segment')['Sales'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data = sales_segment,x ='Segment',y='Sales',palette=('Set1'))
plt.xlabel('Segment')
plt.ylabel('Sales')
plt.title('Sales by Segment')
plt.show()
```
